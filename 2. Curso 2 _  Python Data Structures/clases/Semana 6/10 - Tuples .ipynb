```python
# 3Ordenar por valores instanciados en las llaves
c = {'a':10,'b':1,'c':22}
tmp = list()
for k,v in c.items():
  tmp.append((v,k))
print(tmp)

```

```python
tmp = sorted(tmp, reverse=True)
print(tmp)
```

```python
fhand = open("/content/romeo.txt") #Subes el archivo 
counts = dict() #Creas un diccionario  counts 
for line in fhand: 
  words = line.split() #lees el archivo y con el metodo split, divides las oraciones 
  for word in words:
    counts[word] = counts.get(word,0) + 1 # del mismo modo , como se separo en cadenas, se realiza el conteo si se repite y esto será un key , val
print(counts)
lst = list() # creas una lista nueva 
for key,val in counts.items(): # por cada key value encontrado 
  newtup = (val,key) #se cambia de posicion  val, key
   lst.append(newtup) #se agregar a la lista creada
  print(newtup)

lst = sorted(lst,reverse=True)#Se revierte el orden de manera descendente

for val,key in lst[:10]: # como fue almacenado en una lisata se va imprimir solo desde el0 al 10 de manera descendente
  print(key,val) # su respectivo valor de repticion y palabra
```
