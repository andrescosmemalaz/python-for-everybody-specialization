```python
total = 0 
count = 0 
while True:
  numero = input("Enter your number :")
  if numero == 'done':
    break
  else:
    value = float(numero)
    total = total + value
    count = count + 1

average =  total / count
print('Average : ',average)
```

```python
lista = list()
while True:
  numero = input("Enter your number :")
  if numero == 'done':
    break
  else:
    value = float(numero)
    lista.append(value)

average =  sum(lista) / len(lista)
print(lista)
print('Average : ',average)
```
