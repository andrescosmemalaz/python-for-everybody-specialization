```python
txt = open("/romeo.txt", "r")
count = 0
for line in txt:
  #Este m√©todo se utiliza para eliminar todos los espacios en blanco iniciales y finales de una cadena. ... En realidad strip() devuelve una copia de la cadena con los caracteres iniciales y finales en blanco eliminados.
    print(line.strip())
    count = count + 1
print(count,"Lines")
```
