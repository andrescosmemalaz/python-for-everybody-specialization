```python
counts = dict()
names = ['luisa','derect','maximo','luis','maximo']

```

```python
for name in names:
    if name not in counts:
      counts[name] = 1 
    else:
      counts[name] = counts[name] + 1
print(counts) 
print(names)
```

```python
# El m√©todo get para diccionarios
if name in counts:
  x = counts[name]
else:
  x = 0 

x = counts.get(name,0)
print(x)
```

```python
counts = dict()
names = ['luisa','derect','maximo','luis','maximo']
for name in names:
  counts[name] = counts.get(name,0) + 1 
print(counts)
```
