```python
# Aqui podríamos decir que el bucle nunca llegar a correr
# Por lo explicado anteriormente 
# 
n = 0 
while n > 0:
  print("zero trip loop")
# La razon es que n no es mayor que 0 
print("Zero trip loop ")
# Pero lo que se debe recastar es la estructura de control while, todo lo que esta 
# dentro de la estructura de control va correr
# Lo que esta fuera es parte de otra asignación, por lo que no forma del todo
# parte de la estructura 
```

```python
print("Notemos el poder del bucle while")
# Aqui notamos que la declaración de n = 1
# Nos dice que deseamos que mientras n sea igual o menor 
# nos va imprimir Zero trip loop  
# Como nos consulta ello 5 veces, nos va imprimer n veces 
n = 1
while n <= 5:
  n  = n + 1  
  print('Zero trip loop')
print("END")
```

```python
def planas(planas,numero_veces):
  planas = str(planas)
  numero_veces = int(numero_veces)
  n = 1 
  while n <= numero_veces:
    n  = n + 1  
    print(planas)
```

```python
# No olvidemos que hemos validado que hemos validado que es un string
planas("Andres",5)
```
