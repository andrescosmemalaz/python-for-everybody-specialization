```python
# Determinamos la variable fruit con el string banana
fruit = 'banana'
# Se debe saber que los caracteres en python comienzan a contar desde el número 0
letter = fruit[0]
# Si deseamos avanzar un poco más 
letter_1 = fruit[1]
print(letter)
print(letter_1)
#DEclaramos una variable n con el valor 3 
n = 3
#Posteriormente determinamos en la variable letter_2
# de acuerdo al indice n - 1 , lo que debe ser
#  3 - 1 
letter_2 = fruit[n - 1]
#Nos debe imprimir el indice 2 que es n
print(letter_2)

print(len(fruit))
#Con la funcioncion len podemos determinar el numeros de caracteres 
```
