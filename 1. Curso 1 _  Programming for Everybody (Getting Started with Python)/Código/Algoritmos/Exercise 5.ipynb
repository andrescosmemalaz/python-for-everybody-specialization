```python
# Se declara una variable con un string 
fruit = 'banana'
# iniciamos el lector con el index en 0
index = 0
# Definimos una estructura iterativa 
# donde mientras el index sea menor que 
# la cantidad de letras que encuentra la funci√≥n len
# almacenas el indice y de acuerdo a la palabra , se determina una variable
# letter , conforme se generan las iteraciones, se realiza el analisis 

while index < len(fruit):
    letter = fruit[index]
    print(index, letter)
    index += 1

print (" ")

for letter in fruit:
    print(letter)
```
