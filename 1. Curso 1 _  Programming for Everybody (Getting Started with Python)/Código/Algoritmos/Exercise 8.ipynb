```python

hourly_rate = input("Hourly rate: ")
hourly_rate_int = float(hourly_rate)

hours = input("Number of hours: ")
hours_int = float(hours)


def computepay(h, r):
    overtime_r = r * 1.5

    if h < 40:
        return h * r
    else:
        return (40 * r) + (h - 40) * overtime_r

p = computepay(hours_int, hourly_rate_int)
print(p)
```
