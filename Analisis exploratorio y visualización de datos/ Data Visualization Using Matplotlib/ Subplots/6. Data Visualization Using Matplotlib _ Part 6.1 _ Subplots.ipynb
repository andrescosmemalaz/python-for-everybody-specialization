We can draw subplots in matplotlib using two ways.
# ***1. Using Subplot() method***

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Python/Curso de Técnicas de Operaciones /coursera-python-for-everybody-specialization/Analisis exploratorio y visualización de datos/ Data Visualization Using Matplotlib/ Subplots/Iris.csv')
```

```python
data.head()
```

# ***Dibujar dos Plots side by side (Histogram of SepalLenght & PetalLenght***

```python
data.columns
```

```python
plt.subplot(1,2,1)#1
plt.subplot(1,2,2)#2
```

```python
plt.subplot(1,2,1)#1
plt.hist(data.SepalLengthCm)
plt.subplot(1,2,2)#2
plt.hist(data.PetalLengthCm)
```

```python
plt.suptitle('Iris Dataset')
plt.subplot(1,2,1)#1
plt.title('SepalLengthCm')
plt.hist(data.SepalLengthCm,color ='green')
plt.xlabel('SepalLengthCm')
plt.ylabel('Count')

plt.subplot(1,2,2)#2
plt.title('PetalLengthCm')
plt.hist(data.PetalLengthCm)
plt.xlabel('PetalLengthCm')
plt.ylabel('Count')
plt.show()
```

# ***2. Draw Two plots abajo uno de otro***

```python
plt.figure(figsize=(10,6),edgecolor='black',linewidth=4,facecolor='lightyellow')

plt.suptitle('Iris Dataset')

plt.subplot(2,1,1)#1
plt.title('SepalLengthCm')
plt.hist(data.SepalLengthCm,color ='green')
plt.xlabel('SepalLengthCm')
plt.ylabel('Count')

plt.subplot(2,1,2)#2
plt.title('PetalLengthCm')
plt.hist(data.PetalLengthCm)
plt.xlabel('PetalLengthCm')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
