````python
peliculas = ['Comedia','Terror','Ficci√≤n','Animada']
votos = [4,7,8,9]
````

````python
import matplotlib.pyplot as plt
import numpy as np
````

# ***1. Draw A Pie Chart***

````python
plt.pie(votos)
````

````python
plt.pie(votos)
plt.show()
````

# ***2. Give Title, Label (Also Change Label Distance), And Legend ***

````python
plt.title('Favorite type movie ')
plt.pie(votos)
plt.show()
````

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas)
plt.show()
````

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas)
plt.legend()
plt.show()
````

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1)
plt.legend(loc=0)
plt.show()
````

# ***3. Pull out Action and Drama Wedges***

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0.15,0,0])    
plt.legend(loc=5)
plt.show()
````

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0,0.15])    
plt.legend(loc=5)
plt.show()
````

# ***4. Change The Start Angle To 90 Degree***

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90)    
plt.legend(loc=5)
plt.show()
````

# ***5. Set The Color of Each Wedge***



```
'g' - Green
'r' - red
'y' - yellow
'c' - cyan
'm' - magenta
'w'- White
'k' - Black
'b' - blue
```



````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','k'])    
plt.legend(loc=5)
plt.show()
````

# ***6. Display Percentage & Change Distance Inside Each Wedge***

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','y'],
        autopct = '%0.2f')    
plt.legend(loc=5)
plt.show()
````

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','y'],
        autopct = '%0.2f%%')    
plt.legend(loc=5)
plt.show()
````

````python
plt.title('Favorite type movie ')
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','y'],
        autopct = '%0.2f%%',
        pctdistance= 0.8)    
plt.legend(loc=5,bbox_to_anchor=(1, 0.1, 0.5, 0.5))
plt.show()
````

# ***`7. Change Radius of The Pie Chart`***

````python
plt.title('Favorite type movie ',y=1.35)
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','y'],
        autopct = '%0.2f%%',
        pctdistance= 0.8,
        radius=2)    
plt.legend(loc=5)
plt.show()
````

# ***8. Set Labels Size of The Pie Chart***

````python
plt.title('Favorite type movie ',y=1.35)
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','y'],
        autopct = '%0.2f%%',
        pctdistance= 0.8,
        radius=2,
        textprops={'fontsize':12})    
plt.legend(loc=5,bbox_to_anchor=(1, 1, 0.5, 0.5))
plt.show()
````

# ***9. Set Border of The Wedges***

````python
plt.title('Favorite type movie ',y=1.35)
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','y'],
        autopct = '%0.2f%%',
        pctdistance= 0.8,
        radius=2,
        textprops={'fontsize':12},
      ###
        wedgeprops={'edgecolor':'black',
                    'linewidth':'2',
                    'linestyle':':'},
        shadow = True)   
### 
plt.legend(loc=5)
plt.show()
````

# ***10. Add A shadow to the Pie Chart***

````python
plt.figure(figsize=(10,4),edgecolor='black',
           linewidth= 3,facecolor='lightblue')
plt.title('Favorite type movie ',y=1.35)
plt.pie(votos,labels=peliculas,
        labeldistance=1.1,
        explode= [0,0,0.1,0.1],
        startangle=90,
        colors = ['g','m','c','y'],
        autopct = '%0.2f%%',
        pctdistance= 0.8,
        radius=2,
        textprops={'fontsize':12},
      ###
        wedgeprops={'edgecolor':'black',
                    'linewidth':'2',
                    'linestyle':':'},
        shadow = True)   
### 
plt.legend(loc='upper right',bbox_to_anchor=(1, 1, 0.5, 0.5))
plt.show()
````
