```python
temp= [14.2,16.4,11.9,15.2,18.5,22.1,19.4]
icecream_sales=[215,325,185,332,406,522,412]
tea_sales=[80,65,89,70,56,45,43]
```

```python
import matplotlib.pyplot as plt
```

Questions:

# 1. Draw A Scatter Plot With Title, Xlabel, And YLabel

```python
plt.scatter(temp,icecream_sales)
```

```python
plt.scatter(temp,icecream_sales)
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# **`*2. Change Size of The-Dots*`**

```python
plt.scatter(temp,icecream_sales, s=100)
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# ***3. Change Color of The-Dots***

```python
plt.scatter(temp,icecream_sales, s=100, c = 'red')
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# ***4. Change The Marker***

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(temp,icecream_sales, s=100, c = 'red',marker="D")
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# ***5. Change The Edge Color of The Marker Also change Line Width***

```python
plt.scatter(temp,icecream_sales, s=100, c = 'red',marker="o",
            edgecolors='k',linewidths=2)
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# **6. Change Alpha Parameter**

```python
plt.scatter(temp,icecream_sales, s=100, c = 'red',marker="o",
            edgecolors='k',linewidths=2,
            alpha=0.75 #range0-1)
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# ***7. Set Different Colors For Different Dots***

```python
color = [x for x in range(11,18)]
color
```

```python
plt.scatter(temp,icecream_sales, s=100, c = color ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=0.75) #range0-1
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# ***8. Use In-built Color Maps***

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.scatter(temp,icecream_sales, s=100, c = color ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=0.75,#range0-1
            cmap='Oranges') 
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.show()
```

# ***9. Add Colorbar ***

```python
plt.scatter(temp,icecream_sales, s=100, c = color ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=0.75,#range0-1
            cmap='Oranges') 
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.colorbar()
plt.show()
```

# ***10. Customize The Limit of Colorbar ***

```python
plt.scatter(temp,icecream_sales, s=100, c = color ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=0.75,#range0-1
            cmap='Oranges',
            vmin=10,vmax=20) 
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
c_bar = plt.colorbar()
c_bar.set_label('Temp Vs. Ice Cream Sales')
plt.show()
```

# ***11. Set Different Size For Different Dots***

```python
import numpy as np
size = np.random.randint(low=500,high=2000,size=7)
size
```

```python
plt.scatter(temp,icecream_sales, s=size, c = color ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=0.75,#range0-1
            cmap='Oranges',
            vmin=10,vmax=20) 
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
c_bar = plt.colorbar()
c_bar.set_label('Temp Vs. Ice Cream Sales')
plt.show()
```

# ***12. Draw Two Scatter Plots In One Figure***

```python
plt.scatter(temp,icecream_sales, s=100, c = 'green' ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=1) 
plt.scatter(temp,tea_sales, s=100, c = "red" ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=1)
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
c_bar = plt.colorbar()
c_bar.set_label('Temp Vs. Ice Cream Sales')
plt.show()
```

# ***13. Add Legend To The Plot***

```python
plt.scatter(temp,icecream_sales, s=100, c = 'green' ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=1,label='Ice Cream Sales') 
plt.scatter(temp,tea_sales, s=100, c = "red" ,marker="o",
            edgecolors='k',linewidths=2,
            alpha=1, label = 'Tea Sales')
plt.title('Temp Vs. Ice Cream Sales')
plt.xlabel('Temp')
plt.ylabel('Ice Cream Sales')
plt.legend()
# c_bar = plt.colorbar()
# c_bar.set_label('Temp Vs. Ice Cream Sales')
plt.show()
```
