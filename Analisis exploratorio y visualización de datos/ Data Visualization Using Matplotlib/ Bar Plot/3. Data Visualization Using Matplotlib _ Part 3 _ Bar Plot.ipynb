

```
Grades:  A    B   C    D

```



````python
grades = ['A','B','C','D']
test1=[3,13,10,4]
test2=[5,14,6,5]
test3=[7,15,5,3]
````

````python
import matplotlib.pyplot as plt
import numpy as np
````

# ***1. Draw A Bar Plot***

````python
plt.bar(grades,test1)
````

````python
plt.bar(grades,test1)
plt.show()
````

# ***2. Give Title, Xlabel, and Ylabel ***

````python
plt.bar(grades,test1)
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('No of Students')
plt.show()
````

# ***3. Change Font Size of Title, Xlabel, and Ylabel***

````python
plt.bar(grades,test1)
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.show()
````

# ***4. Set Width of The Bars & Align Bars To Center***

````python
plt.bar(grades,test1,width = 0.6, align='center')
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.show()
````

# ***5. Change Color of The Bars & Also Change Edges Color***

````python
plt.bar(grades,test1,width = 0.6, align='center',
        color ='y')
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.show()
````

````python
plt.bar(grades,test1,width = 0.6, align='center',
        color ='y', edgecolor = 'black')
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.show()
````

````python
plt.bar(grades,test1,width = 0.6, align='center',
        color =['y','b','r','g'])
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.show()
````

# ***6. Change  Line Styles & Line Width of The Bars***

````python
plt.bar(grades,test1,width = 0.6, align='center',
        color ='y', edgecolor = 'black',linestyle=':')
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.show()
````

````python
plt.bar(grades,test1,width = 0.6, align='center',
        color ='y', edgecolor = 'black',linestyle=':',lw=4)
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.show()
````

# ***7. Draw Bar Plots For Test1 and Test2***

````python
index = np.arange(len(grades))
index
````

````python
plt.bar(index,test1,width = 0.4, align='center',
        color ='y', edgecolor = 'black',linestyle=':',lw=4)
plt.title('Bar Plot',fontsize=15)
plt.bar(index+0.4,test2,width = 0.4, align='center',
        color ='g', edgecolor = 'black',linestyle=':',lw=4)
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.xticks(index+0.2,grades)
plt.show()
````

````python
N = 45
x, y = np.random.rand(2, N)
c = np.random.randint(1, 5, size=N)
s = np.random.randint(10, 220, size=N)

fig, ax = plt.subplots()

scatter = ax.scatter(x, y, c=c, s=s)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(), loc="lower left", title="Classes")
ax.add_artist(legend1)

# produce a legend with a cross section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes")

plt.show()
````

# ***9. Add Legend ***

````python
import matplotlib.pyplot as plt
import numpy as np
plt.bar(index,test1,width = 0.3, align='center',
        color ='b', edgecolor = 'black',linestyle=':',lw=4,label='Test 1')
plt.title('Bar Plot',fontsize=15)
plt.bar(index+0.3,test2,width = 0.3, align='center',
        color ='g', edgecolor = 'black',linestyle=':',lw=4,label='Test 2')
plt.title('Bar Plot',fontsize=15)
plt.bar(index+0.6,test3,width = 0.3, align='center',
        color ='y', edgecolor = 'black',linestyle=':',lw=4,label='Test 3')
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.xticks(index+0.3,grades)
plt.legend()
plt.show()
````

# ***10. Change Figure Size & Color***

````python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,3),facecolor='lightblue',edgecolor='black',linewidth=2)
plt.bar(index,test1,width = 0.3, align='center',
        color ='b', edgecolor = 'black',linestyle=':',lw=4,label='Test 1')
plt.title('Bar Plot',fontsize=15)
plt.bar(index+0.3,test2,width = 0.3, align='center',
        color ='g', edgecolor = 'black',linestyle=':',lw=4,label='Test 2')
plt.title('Bar Plot',fontsize=15)
plt.bar(index+0.6,test3,width = 0.3, align='center',
        color ='y', edgecolor = 'black',linestyle=':',lw=4,label='Test 3')
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('No of Students',fontsize=10)
plt.xticks(index+0.3,grades,rotation=60)
plt.legend()
plt.show()
````
