```python
import  matplotlib.pyplot as plt
import numpy as np
```

# ***1. Draw A Line Plot***

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y)
plt.show()
```

# ***2. Give Title, Xlabel and Ylabel ***

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

# ***3. Change Font Size of Title, Xlabel and Ylabel And Also Change location***

```python
import  matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20)
plt.ylabel("Y-Axis",fontsize=20)
plt.show()
```

# ***4. Change Color & Width of The Line***

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y,color='r',linewidth=4)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20)
plt.ylabel("Y-Axis",fontsize=20)
plt.show()
```

# ***5. Change Marker, Marker Size &  Line Style***

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y,color='r',linewidth=4,markersize=10,markerfacecolor='blue',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20)
plt.ylabel("Y-Axis",fontsize=20)
plt.show()
```

# ***6. Add Legend To Your Graph Also Change Location***

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,y,linewidth=4,markersize=10,markerfacecolor='blue',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20)
plt.ylabel("Y-Axis",fontsize=20)
plt.legend(["A simple line plot"],loc=0)
plt.show()
```

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,y,'r--',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.legend(["A simple line plot"],loc=0)
plt.show()
```

# ***7. Create Separate Lines For x and y And Change Plot Style***

```python
plt.style.available
plt.style.use('seaborn')

x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20)
plt.ylabel("Y-Axis",fontsize=20)
plt.legend()
plt.show()
```

# ***9. Change X and Y limit Using axis Method***

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue')
plt.plot(y,'b--',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20)
plt.ylabel("Y-Axis",fontsize=20)
# xmin, xmax, ymin, ymax 
plt.axis([0,20,0,15])
plt.legend(["A simple line plot"],loc=0)
plt.show()
```

# ***10. Change X and Y limit Using Xticks And Yticks.***

```python
x_ticks=[x for x in range(0,20,2)]
y_ticks=[y for y in range(0,15,3)]
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20)
plt.ylabel("Y-Axis",fontsize=20)
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```
