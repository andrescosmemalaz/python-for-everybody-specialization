```python
import re
x = 'My 2 favorites numbers are 19 and 42'
y = re.findall('[0-9]+',x)
for x in y :
  print(x)
# Por expresión regulaes, importando la 
# libreria re , con el método findall 
# nos extrae  el rango exigido, de x 
```

```python
y = re.findall('[AEIOU]+',x)
print(y)
#Como no existe ninguna vocal con mayusculas
# El detalle es mostrado
```

```python
z = "From Using the: character"
# Coincidencia codiciosa
y = re.findall('^F.+:',z)
print(y)
```

```python
z = "From: Using the: character"
# Coincidencia no codiciosa
y = re.findall('^F.+?:',z)
print(y)
```

```python
z = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
# Coincidencia no codiciosa
y = re.findall('\S+@\S+',z)
print(y)
```

```python
z = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
# Coincidencia no codiciosa
y = re.findall('^From (\S+@\S+)',z)
print(y)
```

```python
data = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
atpos = data.find('@')
print(atpos)
```

```python
data = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
sppos = data.find(' ',atpos)
print(sppos)
```

```python
data = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
host = data[atpos+1: sppos]
print(host)
#Sin usar expresiones regulares
```

```python
#Método expresiones regulares
import re
lin = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
y = re.findall('@([^ ]*)',lin)
print(y)
 
```

```python
#Método expresiones regulares
import re
lin = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
y = re.findall('^From .*@([^ ]*)',lin)
print(y)
 
```

```python
Python Regular Expression Quick Guide

^        Matches the beginning of a line
$        Matches the end of the line
.        Matches any character
\s       Matches whitespace
\S       Matches any non-whitespace character
*        Repeats a character zero or more times
*?       Repeats a character zero or more times 
         (non-greedy)
+        Repeats a character one or more times
+?       Repeats a character one or more times 
         (non-greedy)
[aeiou]  Matches a single character in the listed set
[^XYZ]   Matches a single character not in the listed set
[a-z0-9] The set of characters can include a range
(        Indicates where string extraction is to start
)        Indicates where string extraction is to end
```
