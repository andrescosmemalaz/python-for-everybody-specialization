Ya has visto en la clase que puedes cortar una cadena con una llamada como s[i:j], que te da una porción de la cadena s desde el índice i hasta el índice j-1. Sin embargo, esta no es la única manera de cortar una cadena. Si se omite el índice inicial, Python asumirá que se desea comenzar el corte en el índice 0. Si omite el índice final, Python asumirá que desea terminar su rebanada al final de la cadena. Comprueba esta sesión con el shell de Python:

````python
s = 'Python is Fun!'
````

````python
s[1:5]
````

````python
s[:5]
````

````python
s[1:]
````

````python
s[:]
````

El último ejemplo es interesante. Si se omite el índice inicial y el final, se obtiene la cadena original.

Hay otra cosa interesante que puedes hacer con el corte de cadenas. Puedes añadir un tercer parámetro, k, así: s[i:j:k]. Esto da una rebanada de la cadena s desde el índice i hasta el índice j-1, con un tamaño de paso k. Echa un vistazo a los siguientes ejemplos:

````python
s = 'Python is Fun!'
````

````python
s[1:12:2]
````

````python
s[1:12:3]
````

````python
s[::2]
````

El último ejemplo es similar al ejemplo s[:]. Con s[::2], estamos pidiendo la cadena completa s (desde el índice 0 hasta el 13), con un tamaño de paso de 2 - por lo que terminamos con cada uno de los caracteres de s. ¡Muy bien!
