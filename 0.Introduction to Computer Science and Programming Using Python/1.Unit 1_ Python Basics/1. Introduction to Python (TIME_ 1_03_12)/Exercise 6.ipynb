Para el último problema (10), escribir la expresión en su intérprete de Python puede dar un resultado que no es exacto. Por ejemplo, 6,6000000000000005 en lugar de 6,6. Esto se debe a que los ordenadores tienen dificultades para almacenar las fracciones exactamente en binario. Así que cuando se realizan cálculos, los números se convierten primero a binario, luego se realiza el cálculo y el resultado se convierte de nuevo en decimal. Sin embargo, las fracciones no pueden almacenarse exactamente en binario (por ejemplo, no hay forma de almacenar 0,33333 repitiendo exactamente) por lo que se introducen estos pequeños errores de redondeo que se propagan hasta la respuesta final.
