En este problema, crearás un programa que adivine un número secreto.

El programa funciona de la siguiente manera: usted (el usuario) piensa en un número entero entre 0 (inclusive) y 100 (no inclusive). El ordenador hace conjeturas, y tú le das información: ¿es su conjetura demasiado alta o demasiado baja? Utilizando la búsqueda por bisección, el ordenador adivinará el número secreto del usuario.

A continuación, la transcripción de una sesión de ejemplo:

```python
print('Please think of a number between 0 and 100!')
lo, hi, ans = 0, 100, None

while ans != 'c':
    guess = (lo + hi) // 2
    print('Is your secret number {}?'.format(guess))
    ans = input("Enter h, l or c for hi, lo or correct guesses: ")
    if ans == 'h':
        hi = guess
    elif ans == 'l':
        lo = guess
    elif ans != 'c':
        print('Sorry, I did not understand your input.')
print('Game over. Your secret number was: ', guess)
```
