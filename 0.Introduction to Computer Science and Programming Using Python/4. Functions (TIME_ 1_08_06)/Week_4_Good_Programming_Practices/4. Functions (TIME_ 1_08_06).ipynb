```python
def d(x, y):
   '''
   x: Can be int or float.
   y: Can be int or float.
   '''
   return x > y
```

```python
def e(x, y, z):
   '''
   x: Can be int or float.
   y: Can be int or float.
   z: Can be int or float.
   '''
   return x >= y and x <= z
 
```

```python
e(15,20,15)
```

```python
def f(x, y):
   '''
   x: int or float.
   y: int or float
   '''
   x + y - 2
```

```python
f(20,15)
```

Explicación:

La última función, a diferencia de las otras, no tiene declaración de retorno. Sólo realiza una operación (la operación es x+y-2). Como no devuelve nada explícitamente, Python devuelve por defecto el valor None cuyo tipo es NoneType. Así que esta función y cualquier otra que no tenga una sentencia de retorno se puede reescribir como:

```python
def f(x, y):
   x + y - 2
   return None
      
```

```python
f(15,20)
```

TIEMPO ESTIMADO PARA TERMINAR: 5 minutos

Escribe una función en Python, evalQuadratic(a, b, c, x), que devuelva el valor de la cuadrática .

Esta función toma cuatro números y devuelve un único número.

```python
def evalQuadratic(a, b, c, x):
  valor = a * (x * x) + (b * x) + c  
  print(valor)
```

```python
evalQuadratic(2,8,7,4)
```

```python
def evalQuadratic(a, b, c, x):
   '''
   a, b, c: numerical values for the coefficients of a quadratic equation
   x: numerical value at which to evaluate the quadratic.
   '''
   return a*x*x + b*x + c
```

```python
evalQuadratic(2,8,7,4)
```
