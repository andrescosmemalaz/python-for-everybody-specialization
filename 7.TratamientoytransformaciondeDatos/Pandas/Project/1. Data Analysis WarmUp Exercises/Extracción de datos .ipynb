```python
import re
def get_geocoder(url_location):
  try:
    coords = re.search(r"!3d-?\d\d?\.\d{4,8}!4d-?\d\d?\.\d{4,8}",
                     url_location).group()
    coord = coords.split('!3d')[1]
    return list(coord.split('!4d'))
  except (TypeError, AttributeError):
            return ("", "")
```
