# ***22. Find Unique Values From Genre ***

```python
data['Genre']
```

```python
list1 =  list()
for x in data['Genre']:
  list1.append(x.split(","))
  
```

```python
list1
```

```python
one_word = list()
for item in list1:
  for i1 in item:
    one_word.append(i1)
```

```python
one_word
```

```python
len(one_word)
```

```python
unica_list=list()
for item in one_word:
  if item not in unica_list:
    unica_list.append(item)
```

```python
unica_list
```

```python
len(unica_list)
```
